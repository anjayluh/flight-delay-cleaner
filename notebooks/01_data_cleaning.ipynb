```python
import pandas as pd

df = pd.read_csv("../data/raw/flights_raw.csv")
```

```python
df = df.drop_duplicates()
```

```python
print(df.columns.to_list())
```

```python
# B. Convert date/time columns
# Create a proper date column
df['FL_DATE'] = pd.to_datetime(df[['YEAR', 'MONTH', 'DAY']], errors='coerce')

# Check it
df[['YEAR','MONTH','DAY','FL_DATE']].head()
```

```python
# C. Fix missing values

# Numeric columns:

num_cols = df.select_dtypes(include=['float64','int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())


# Categorical columns:

cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

```python
print(df.columns.to_list())
```

```python
# D. Remove negative or impossible values
df = df[df['DEPARTURE_DELAY'] >= 0].copy()
df = df[df['ARRIVAL_DELAY'] >= 0].copy()
```

```python
# E. Handle outliers

# Use IQR:

Q1 = df['ARRIVAL_DELAY'].quantile(0.25)
Q3 = df['ARRIVAL_DELAY'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['ARRIVAL_DELAY'] >= Q1 - 1.5*IQR) & 
        (df['ARRIVAL_DELAY'] <= Q3 + 1.5*IQR)]
```

```python
# STEP 7 — Feature Engineering

# Extract useful features from dates:

df['day_of_week'] = df['FL_DATE'].dt.dayofweek
df['month'] = df['FL_DATE'].dt.month
df['is_weekend'] = df['day_of_week'] >= 5


# Convert boolean to int:

df['is_weekend'] = df['is_weekend'].astype(int)


# One-hot encode categorical variables:

df = pd.get_dummies(df, columns=['AIRLINE', 'ORIGIN_AIRPORT', 'DESTINATION_AIRPORT'], drop_first=True)
```

```python
# STEP 8 — Save the Cleaned Dataset
df.to_csv("../data/cleaned/flights_cleaned.csv", index=False)


# This file is gold — employers love seeing before/after datasets.

# Optional: STEP 9 — Build a Tiny ML Model
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = df.drop("ARRIVAL_DELAY", axis=1)
y = df["ARRIVAL_DELAY"] > 15  # delayed or not

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

print(model.score(X_test, y_test))


# Save the model:

import joblib
joblib.dump(model, "../models/baseline_model.pkl")
```
